# Areas Criteria
# Normality for Read Variable
1. Area between the curve mean+-std.The no. of counted values between x1 and x2 is 130 and the total values is 200 so the are is 130/200=0.65
2. Area between the curve mean+-2std and this area should be greater than or equal to 0.955. THe counted values are 192 therefore the area is 192/200=0.96
3. Calculate the area under the curve mean+-3std and this should be greater than 0.997 .the number of counted values is 200 therefore the area of the curve is 200/200=1

#### In the variable read the first criteria is not satisfied but the remaining two criterias are satisfied.For a variable to have a normal distribution all the three criterias should be satisfied.Therefore we can conclude that the variable read doesn't have a normal distribution or it is very close to being normally distributed since two criterias are satisfied.

# Test of Outliers in the Data:-

#### Outliers are the extreme values in the data.To find the outliers the statistics used are Q1 and Q3
#### step 1.Criteria 1- Calculate the value of IQR=Q3-Q1
#### step 2.Criteria 2-Calculate the value of left cut-off=Q1-1.5*IQR
#### step 3.To get the right cut-off the formula is Q3+1.5*IQR
#### All the datapoints less than criteria 1 and greater than criteria 2 will be counted as an outlier.
#### Calculate the no. of outliers from both the criteria and then report the total number of outliers.
## Lesser the outliers ,better the data.
